```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
f = pd.read_csv('air_traffic.csv')
f
```

```python
plt.figure(figsize=(16, 8))
plt.plot(f['volume'])
plt.show()
```

```python
tau = 3
S = []
for i in range(tau, len(f) - tau):
	S.append(np.log(f['volume'][i - tau] * f['volume'][i + tau] / f['volume'][i]**2))
```

```python
plt.figure(figsize=(16, 8))
plt.plot(S)
plt.show()
```
